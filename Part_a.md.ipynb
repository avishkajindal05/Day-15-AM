# Probability Distributions – Real World Scenarios

This notebook identifies distributions, explains reasoning, and computes probabilities using scipy.stats.

```python
import numpy as np
from scipy import stats
```

## 1. Website Traffic
Distribution: Poisson
Reasons:
- Counts number of events in fixed interval
- Events occur independently with constant average rate


```python
lam = 200
prob_more_220 = 1 - stats.poisson.cdf(220, lam)
prob_more_220
```

## 2. Quality Control (Defective Bolts)
Distribution: Binomial
Reasons:
- Fixed number of trials (50 bolts)
- Each bolt defective with constant probability 0.02


```python
prob_3_defective = stats.binom.pmf(3, 50, 0.02)
prob_3_defective
```

## 3. Delivery Times
Distribution: Normal N(45, 8^2)
Reasons:
- Continuous measurement
- Influenced by many small independent factors


```python
prob_gt_60 = 1 - stats.norm.cdf(60, 45, 8)
prob_40_50 = stats.norm.cdf(50,45,8) - stats.norm.cdf(40,45,8)
prob_gt_60, prob_40_50
```

## 4. Customer Arrivals
Distribution: Poisson
Reasons:
- Counts number of arrivals in a time interval
- Independent arrivals with constant rate


```python
lam = 10/10
prob_zero = stats.poisson.pmf(0, lam)
prob_zero
```

## 5. Class Average Score (CLT)
Distribution: Approximately Normal
Reasons:
- Mean of many independent observations
- Central Limit Theorem applies for large samples


```python
mean = 70
sd = 10
n = 35
sampling_sd = sd/np.sqrt(n)
sampling_sd
```
