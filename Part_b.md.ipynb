# Beta Distribution Analysis with Bayesian Updating

This notebook demonstrates:
1. Plotting PDFs for Beta(2,5), Beta(5,5), Beta(0.5,0.5)
2. Interpreting these shapes as prior beliefs about a coin's bias
3. Bayesian updating when observing 7 heads in 10 flips starting from a Beta(1,1) prior

Libraries used: `numpy`, `matplotlib`, `scipy.stats`.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

## 1. Plot Beta Distribution PDFs

We plot three Beta distributions:

- Beta(2,5) → belief that coin is biased toward tails
- Beta(5,5) → belief coin is fair
- Beta(0.5,0.5) → belief coin is strongly biased but unknown direction

```python
x = np.linspace(0, 1, 500)

pdf_25 = beta.pdf(x, 2, 5)
pdf_55 = beta.pdf(x, 5, 5)
pdf_05 = beta.pdf(x, 0.5, 0.5)

plt.figure()
plt.plot(x, pdf_25, label='Beta(2,5)')
plt.plot(x, pdf_55, label='Beta(5,5)')
plt.plot(x, pdf_05, label='Beta(0.5,0.5)')

plt.title('Beta Distribution PDFs')
plt.xlabel('Probability of Heads')
plt.ylabel('Density')
plt.legend()
plt.show()
```

## 2. Interpretation as Prior Beliefs

### Beta(2,5)
- Mean = 2 / (2+5) ≈ 0.286
- Prior belief that the coin lands tails more often than heads

### Beta(5,5)
- Mean = 0.5
- Strong belief the coin is fair

### Beta(0.5,0.5)
- U-shaped distribution
- Belief that the coin is likely strongly biased (either mostly heads or mostly tails)

## 3. Bayesian Updating Example

Start with a **Beta(1,1)** prior (uniform).

Observe:
- 7 heads
- 3 tails

Posterior formula:

Posterior = Beta(α + heads, β + tails)

So:

Beta(1+7, 1+3) = **Beta(8,4)**

```python
x = np.linspace(0, 1, 500)

prior = beta.pdf(x, 1, 1)
posterior = beta.pdf(x, 8, 4)

plt.figure()
plt.plot(x, prior, label='Prior Beta(1,1)')
plt.plot(x, posterior, label='Posterior Beta(8,4)')

plt.title('Bayesian Update for Coin Bias')
plt.xlabel('Probability of Heads')
plt.ylabel('Density')
plt.legend()
plt.show()
```

## Posterior Mean

Posterior mean for Beta(8,4):

8 / (8+4) = **0.667**

Interpretation:

After observing 7 heads in 10 flips, our belief shifts toward the coin having roughly **66–67% probability of heads**.
